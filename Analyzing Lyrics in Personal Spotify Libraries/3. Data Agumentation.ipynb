## 3.0 Create our TOKENS and VOCABULARY tables

## 3.1 Now that we know what our text looks like we have an idea of how we will parse it

```python
line_pat = r"\\n" # How we will separate each line
section_pat = r"\[[\s\w\:\w\-\s\"\\(\)']+\]" # How we will define and separate each section of the songs
```

## 3.2 Chunk by Section
We will only be working with the texts in English

```python
SECTION = CORPUS_english['lyrics'].str.split(section_pat, expand=True).stack().reset_index(level=0, drop=True).to_frame('section_str')
SECTION['song_num'] = (SECTION.index==0).cumsum() -1
SECTION['track.id']  = CORPUS_english.loc[list(SECTION['song_num'])]['track.id'].to_frame().set_index(SECTION.index)
SECTION = SECTION.drop('song_num',axis=1)
SECTION = SECTION.set_index(['track.id',SECTION.index])
SECTION.index.names = ['track.id','section_num']
SECTION = SECTION.drop(SECTION[SECTION['section_str'] == ''].index)
SECTION[:20]
```

## 3.3 Chunk by Line

```python
LINES = SECTION['section_str'].str.split('\n', expand=True).stack().to_frame('line_str')
LINES = LINES.drop(LINES[LINES['line_str'] ==''].index)
LINES.index.names = ['track.id','section_num','line_num']
LINES[:20]
```

## 3.4 Remove punctuations from our lines so that they are not tokenized

```python


def keep_alpha_numeric(string):
    pattern = re.compile(r'[^a-zA-Z0-9\s]')
    return re.sub(pattern, '', string)

LINES['line_str'] = LINES['line_str'].apply(keep_alpha_numeric)
LINES[:30]
```

## 3.5 Remove stop words from our lines

It is important to note here that if we would like to compare styles and not just content between artists or within their discography, it would be important to keep stop words

## 3.6 Generate our `TOKENS` table 

```python
# There may be more elegant ways to do this but this was how I was able to track its progress and it happened to run faster than
# some comprehension methods. I'd like to look into why at a later time. 
TOKENS = pd.DataFrame()
song_count = 0
for song, group_df in LINES.groupby(level=0):
    if song_count%100 ==0:
        print(f"Completed {song_count} songs.")
    df = group_df['line_str'].apply(lambda x: pd.Series(nltk.pos_tag(nltk.word_tokenize(x)),dtype=str))
    df = df.stack()
    df = df.to_frame('pos_tuple')
    
    TOKENS = pd.concat([TOKENS,df])
    song_count +=1 

    
```

```python
TOKENS.index.names = OHCO
TOKENS['pos'] = TOKENS.pos_tuple.apply(lambda x: x[1])
TOKENS['token_str'] = TOKENS.pos_tuple.apply(lambda x: x[0])
TOKENS['term_str'] = TOKENS.token_str.str.lower()
TOKENS
```

```python
TOKENS.to_csv(data_dir+'TOKENS_english.csv') #save our tokens 
```

## 3.5 Generate our `VOCAB` table

```python
VOCAB = TOKENS.term_str.value_counts().to_frame('n')
VOCAB.index.name = 'term_str'
VOCAB['p'] = VOCAB.n / VOCAB.n.sum()
VOCAB['i'] = -np.log2(VOCAB.p)
VOCAB['n_chars'] = VOCAB.index.str.len()
VOCAB['max_pos'] = TOKENS[['term_str','pos']].value_counts().unstack(fill_value=0).idxmax(1)
VOCAB.sort_values('n',ascending=False)
```

```python
VOCAB.to_csv(data_dir+'VOCAB_english.csv')
```
