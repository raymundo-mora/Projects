# 1.1 Quick Data Cleanup

````python
CORPUS['song_id'] = LIB['track.id'] # add the song id to our CORPUS
````

````python
CORPUS = CORPUS.dropna(subset='lyrics') # drop any songs that have missing lyrics
CORPUS
````

````python
LIB = LIB.loc[CORPUS.index].reset_index(drop=True) # rest indeces
CORPUS = CORPUS.reset_index(drop=True)
````

````python
LIB[['artist','song']] = CORPUS[['artist','song']] # add artist and song titles to our library
LIB
````

````python
def first_genre(genres):
    genres = ast.literal_eval(genres)
    if len(genres) == 0:
        return 'NA'
    else:
        return genres[0]

LIB['top_genre'] = LIB['genre'].apply(first_genre)
LIB
````

## 2.0 Language Statistics

````python
counts = CORPUS['language_name'].value_counts()
plt.figure(figsize=(10,6))
plt.bar(counts.index,counts)
plt.xticks(rotation=45)
plt.xlabel('Language')
plt.ylabel('Counts')
plt.suptitle('Language Frequency')
for i, count in enumerate(counts):
    plt.text(i, count+15, str(count), ha='center', fontsize=10)


plt.tight_layout()
plt.savefig(vis_dir+"2.2.1.png",bbox_inches='tight',pad_inches=0.1)

plt.show()
````

## 4.0 Genre Statistics

````python
import ast
all_genres = []
for i in LIB.index:
    gen_list = ast.literal_eval(LIB['genre'][i])
    all_genres.extend(gen_list)
````

````python
gen_list
````

````python
GENRES = pd.DataFrame(columns=['genre','n'])
for item in set(all_genres):
    GENRES.loc[-1] = [item,all_genres.count(item)]  # adding a row
    GENRES.index = GENRES.index + 1  # shifting index
    GENRES = GENRES.sort_index()
````

````python
GENRES = GENRES.sort_values(by='n',ascending=False).reset_index(drop=True)
GENRES
````

````python
GENRES_small = GENRES.sort_values(by='n',ascending=False)[:20]
GENRES_small
````

````python
plt.figure(figsize=(10,6))
plt.barh(GENRES_small.genre,GENRES_small.n)

plt.xlabel('Frequency')
plt.xticks(ha='right')
plt.ylabel('Genre')
plt.suptitle('Top 20 Genres')


for i in GENRES_small.index:
    genre = GENRES_small.iloc[i]['genre']
    genre_count = GENRES_small.iloc[i]['n']
    plt.text(genre_count+15, genre, str(genre_count), ha='center', fontsize=10)
    
plt.tight_layout()
plt.savefig(vis_dir+"2.4.1.png",bbox_inches='tight',pad_inches=0.1)
plt.show()
````
